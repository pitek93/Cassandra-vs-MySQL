```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#TEST1

x = np.array([5,10,15,20])
#relacyjna
y = np.array([11.7366361618042, 23.00187373161316, 35.28084588050842, 46.77485370635986])

#obiektowa
y1 = np.array([14.94456171989441, 18.013745307922363, 21.451774835586548, 24.14399027824402])

plt.figure()
plt.plot(x, y, 'b-', marker = 'x', label="baza relacyjna")
plt.plot(x, y1, 'r-', marker = 'x', label="baza obiektowa")

plt.xlabel("liczba rekordów (ocen do filmów) [mln]")
plt.ylabel("czas wykonywania operacji [s]")
plt.xticks([5,7,10,15,20])

plt.legend()
plt.show()
```
